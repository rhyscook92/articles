```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# given escape times data
escape_times = [160, 30, 90, 60, 15, 28, 20, 30, 22, 11, 15, 20, 12, 10, 14, 10, 8, 8, 5, 10, 8, 6, 6, 7]
# create a DataFrame
df = pd.DataFrame({'Trial': range(1, len(escape_times) + 1), 'Escape Time': escape_times})
```

```python
# set a style and context to make the plot look professional
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.4)

# create the plot
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=df, x='Trial', y='Escape Time', color='darkblue', linewidth=2.5)

# set titles and labels
ax.set_title("Thorndike's Cat Escape Time", fontsize=20, pad=20)
ax.set_xlabel('Trial Number', fontsize=15)
ax.set_ylabel('Escape Time (sec)', fontsize=15)

# save the plot
fig.savefig("escape_time_plot.png")

plt.show()
```

```python
response_times = [5*60, 51.5*60, 47.5*60, 25*60, 43, 21, 28, 10, 10, 15] # Convert first four responses to seconds
response_number = list(range(1, len(response_times)+1))  # Create a list of response numbers

# Create a pandas DataFrame from the data
df = pd.DataFrame({
    'Response_Number': response_number,
    'Response_Time': response_times
})
```

```python
# set a style and context to make the plot look professional
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.4)

# create the plot
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=df, x='Response_Number', y='Response_Time', color='darkblue', linewidth=2.5)

# set titles and labels
ax.set_title("Skinner's Rat Response Time", fontsize=20, pad=20)
ax.set_xlabel('Response Number', fontsize=15)
ax.set_ylabel('Response Time (seconds)', fontsize=15)

# save the plot
fig.savefig("skinner_rat_response.png")

plt.show()
```
